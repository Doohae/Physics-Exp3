```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from typing import List, Optional, Union
```

```python
# Exp Data
radius_I = {
    2.5: [1.62, 1.81, 1.94, 2.05, 2.17, 2.24],
    3: [1.36, 1.46, 1.64, 1.7, 1.85, 1.9, 2.01, 2.09, 2.17, 2.24],
    4: [0.97 ,1.05 ,1.19 ,1.25 ,1.39 ,1.43 ,1.53 ,1.56 ,1.65 ,1.68],
    5: [0.74, 0.8, 0.95, 1, 1.06, 1.13, 1.2, 1.24, 1.29 ,1.34]
}
```

```python
mu_0 = 4*np.pi*1e-7
n = 130
R = 0.15
def get_magnetic(currents: List[float]) -> List[float]:
    magnets = []
    for current in currents:
        magnet = (0.8**1.5)*mu_0*n*current/0.15
        magnets.append(magnet)
    return magnets
```

```python
def charge_to_mass(radius: float, magnets: List[float], volts: List[float]) -> List[float]:
    ratios = []
    for mag, vol in zip(magnets, volts):
        ratio = 2*vol/(mag*radius)**2
        ratios.append(ratio)
    return ratios
```

```python
all_ratio = dict()
all_magnets = dict()
for radius, currents in radius_I.items():
    magnets = get_magnetic(currents)
    all_magnets[radius] = magnets
    volts = [120+20*idx for idx, _ in enumerate(currents)]
    specific = charge_to_mass(radius/100, magnets, volts)
    all_ratio[radius] = specific

all_ratio
```

```python
def error_rate(exp_value: List[float], ground_truth: Union[List[float], float])->List[float]:
    if isinstance(ground_truth, float):
        ground_truth = [ground_truth for _ in range(len(exp_value))]
    return [100*abs(gt-v)/gt for gt, v in zip(ground_truth, exp_value)]
```

```python
all_error = dict()
actual_ratio = 1.84*1e+11
for radius, ratios in all_ratio.items():
    all_error[radius] = error_rate(ratios, actual_ratio)
all_error
```

```python
# Exp. Specific Charge vs. Theor.
%matplotlib inline

xticks = [120+20*idx for idx, _ in enumerate(all_ratio[5])]
actual_ratio = 1.84
for radius, ratios in all_ratio.items():
    plt.plot([120+20*idx for idx, _ in enumerate(ratios)], [ratio/(1e+11) for ratio in ratios])
plt.axhline(y=actual_ratio, color='aqua', linestyle='--')
plt.xlabel("Voltage (V)")
plt.ylabel("Specific Charge ($X10^{11}$)")
plt.xticks(xticks)
plt.ylim([1.4, 3])
plt.title("Specific Charge Exp. Results")
plt.legend(all_ratio.keys())
plt.show()
```
